```python
## Import libraries
from pathlib import Path

import matplotlib.pyplot as plt

from opencropphenotyping.io import (read_band, find_band)
```

```python
## Find paths to the bands of interest

project_root = Path.cwd().parent

safe_dir = (project_root / "data" / "raw" / "sentinel_2" / "S2A_MSIL2A_20250804T104701_N0511_R051_T31TCJ_20250804T161517.SAFE")

b04_path = find_band(safe_dir, "B04")
b08_path = find_band(safe_dir, "B08")

print(b04_path.name)
print(b08_path.name)
```

```python
## Read the bands

red, profile = read_band(b04_path)
nir, _ = read_band(b08_path)

print(red.shape)
print(red.dtype)

print(nir.shape)
print(nir.dtype)

print(profile)
```

```python
## Visualise the bands
fig, ax = plt.subplots(figsize=(8,8))
ax.imshow(red, cmap="gray")
ax.set_title("Band B04 - Red")
#plt.imshow(red, cmap="gray")
#plt.title("Band B04 - Red")
ax.colorbar()
ax.show()

fig, ax = plt.subplots(figsize=(8,8))
ax.imshow(nir, cmap="gray")
ax.set_title("Band B08 - NIR")
#plt.imshow(nir, cmap="gray")
#plt.title("Band B08 - NIR")
ax.colorbar()
ax.show()
```
